# BOOK SALES DATA ANALYSIS - PYTHON - IMPLEMENTATION

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Books_Data_Clean.csv')
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df.info()
```

```python
# Summary of Numerical Columns
df.describe()
```

```python
df = df[df['Publishing Year'] > 1900]
```

```python
# Check Null Values
df.isna().sum()
```

```python
# Removing Empty Book Name Rows
df.dropna(subset='Book Name', inplace = True)
```

```python
# Check Duplicates
df.duplicated().sum()
```

```python
# Number of Unique Entries By Column
df.nunique()
```

```python
# Distribution of Publishing Year
plt.figure(figsize=(10,6))
plt.hist(df['Publishing Year'], color = '#EF4026')
plt.xlabel('Publishing Year')
plt.ylabel('Frequency')
plt.title('Distribution of Publishing Year')
plt.show()
```

```python
# Number of Books in Each Genre
df['genre'].value_counts()
```

```python
plt.figure(figsize=(10,6))
df['genre'].value_counts().plot(kind='bar', color='#EF4026')
plt.xlabel('Genre')
plt.ylabel('Number of Books')
plt.title('Number of Books in Each Genre')
plt.show()
```

```python
# Author By Average Ratings
df.groupby('Author')['Book_average_rating'].mean().sort_values(ascending=False)
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x ='genre', y ='Book_ratings_count', data = df, color = '#EF4026')
plt.xlabel('Genre')
plt.ylabel('Books Ratings Count')
plt.title('Book Ratings Count for Each Genre')
plt.show()
```

```python
# Sales Price by Unit Sold
plt.figure(figsize=(10,6))
plt.scatter(df['sale price'], df['units sold'], color = '#EF4026')
plt.xlabel('Sale Price')
plt.ylabel('Units Sold')
plt.title('Sale Price v/s Units Sold')
plt.show()
```

```python
df['language_code'].value_counts().sort_values(ascending = True)
```

```python
# Pie Chart for Language Code
language_counts = df['language_code'].value_counts()
```

```python
plt.figure(figsize=(10,6))
plt.pie(language_counts, labels = language_counts.index, autopct = "%1.1f%%")
plt.title('Language Distribution of Books')
plt.axis('equal')
plt.show()
```

```python
# Publisher by Revenue
df.groupby('Publisher ')['publisher revenue'].sum().sort_values(ascending = False)
```

```python
# Author Rating and Average Book Rating
df.groupby('Author_Rating')['Book_ratings_count'].mean().sort_values(ascending = False).max()
```

```python
# Language Count by Number of Books in each language
df.groupby('language_code').size().sort_values(ascending = False)
```

```python
# Author Rating by Book Rating count
df.groupby('Author_Rating')['Book_ratings_count'].max()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df['Book_average_rating'], df['Book_ratings_count'], color = '#EF4026')
plt.xlabel('Book Average Rating')
plt.ylabel('Book Ratings Count')
plt.title('Book Average Rating v/s Book Ratings Count')
plt.show()
```

```python
# Total gross sales for each author
total_gross_sales_by_author = df.groupby('Author')['gross sales'].sum()
```

```python
total_gross_sales_by_author.sort_values(ascending = False).head(10)
```

```python
plt.figure(figsize=(10,6))
total_gross_sales_by_author.sort_values(ascending = False).head(20).plot(kind='bar', color ='#EF4026')
plt.xlabel('Author')
plt.ylabel('Total Gross Sales')
plt.title('Total Gross Sales for Each Author')
plt.show()
```

```python
# Unit Sold for Each Author Rating
plt.figure(figsize=(10,6))
sns.boxplot(x ='Author_Rating', y ='units sold', data = df, color = '#EF4026')
plt.xlabel('Author Rating')
plt.ylabel('Units Sold')
plt.title('Units Sold for Each Author Rating')
plt.show()
```

```python
# Units Sold by Publishing Year
plt.figure(figsize=(10,6))
df.groupby('Publishing Year')['units sold'].sum().plot(kind = 'line', color = '#EF4026', marker = 'o')
plt.xlabel('Publishing Year')
plt.ylabel('Total Units Sold')
plt.title('Total Units Sold Over Years')
plt.show()
```
